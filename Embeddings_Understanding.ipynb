**1.what is Embeddings??**

-Embeddings are numerical representations of text, images, audio, or any data. For text (usually our focus) an embedding is:

- a vector (a long list of numbers)

- Generated by a model

- Designed so that similar meaning- similar numbers

**2. Why Do we Use Embeddings??**

- because computers don't understand text directly - they understand numbers,

- Embeddings allow systems to:

- Search semantically ('Meaning-based search')

- Build recommendation systems

- Measure similarity between texts

- Cluster docments

- RAF (Retrival-Augmented Generation), including ChatGroq workflows

**3.What is Similarity Score??**

- Once you have embeddding, you measure similarity using a mathematical formula.

- The most common one is:

-Cosine Simlarity:

-it measures the angle between two vectors.

    -1.0 -> identical meaning

    -0.0 -> unrelated

    -1.0 -> Opposite meaning

```python
v1=embedding[0].reshape(-1,1) # i like apples
v2=embedding[1].reshape(-1,1) # i enjoy eat apple
v3=embedding[2].reshape(-1,1)# i drive a motorecycle
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(v1,v2)
```

```python
def cosine_sim(a,b):
    return np.dot(a,b)/(np.linalg.norm(a) * np.linalg.norm(b))
```

```python
v1=embedding[0]
v2=embedding[1]
cosine_sim(v1,v2)
```

```python
v1=embedding[0]
v3=embedding[2]
cosine_sim(v1,v3)
```

```python
nr=np.dot(v1,v2)
norm_v1=np.linalg.norm(v1)
norm_v2=np.linalg.norm(v2)
dr=norm_v1*norm_v2
cosine_sim(v1,v2),nr/dr
```

```python
nr=np.dot(v1,v3)
norm_v1=np.linalg.norm(v1)
norm_v2=np.linalg.norm(v3)
dr=norm_v1*norm_v2
cosine_sim(v1,v2),nr/dr
```
