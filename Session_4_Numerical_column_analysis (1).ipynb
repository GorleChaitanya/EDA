**Emperical rule**

<img src="https://statisticsbyjim.com/wp-content/uploads/2021/08/empirical_rule_graph2.png" jsaction="" class="sFlh5c FyHeAf iPVvYb" style="max-width: 572px; height: 205px; margin: 0px; width: 305px;" alt="Empirical Rule: Definition &amp; Formula - Statistics By Jim" jsname="kn3ccd">

```python
# u-1*sigma  to  u+1*sigma  = 68
# step-1: means of wage
# step-2: std of wage
# step-3: lb=mean-1*sigma
#         lb= s1 - 1* s2
# step-4: ub=mean+1*sigma
```

```python
wage_data=visa_df['prevailing_wage']
wage_mean=visa_df['prevailing_wage'].mean()
wage_std=visa_df['prevailing_wage'].std()
lb=wage_mean-1*wage_std
ub=wage_mean+1*wage_std
lb,ub
con1=wage_data>lb # u-1*sigma
con2=wage_data<ub # u+1*sigma
con=con1 & con2  # u-1*sigma to u+1*sigma
len(visa_df[con])==68*25480/100
```

```python
# step-1: q1
# step-2: q2
# step-3: q3
# step-4: iqr=q3-q1
# step-5: lb= q1-1.5*iqr
# step-6: ub=q3+1.5*iqr
# step-7: con1
# step-8: con2
# step-9: con
# step-10: get the count
```

```python
q1 = np.percentile(wage_data, 25)
q2 = np.percentile(wage_data, 50)
q3 = np.percentile(wage_data, 75)
iqr = q3 - q1
lb = q1 - 1.5 * iqr
ub = q3 + 1.5 * iqr
con1 = wage_data<lb
con2 = wage_data>ub
con3 = con1 | con2
outliers_df=visa_df[con3]
outliers_df['prevailing_wage']
outliers_df['prevailing_wage'].values
outliers_list=outliers_df['prevailing_wage'].values.tolist()
```

```python
# i want to call the each observation of wage data
# if that observation will avaialble in this 427 values
# the that observation will replace by median
# otherwise that observation will be avalable as it is

wage_median=visa_df['prevailing_wage'].median()
wage_data=visa_df['prevailing_wage']
new_wage_data=[]
for i in wage_data:
  if i in outliers_list:
      new_wage_data.append(wage_median)
  else:
    new_wage_data.append(i)
```

```python
plt.figure(figsize=(10,6))
plt.subplot(2,2,1)
plt.hist(wage_data)
plt.subplot(2,2,2)
plt.hist(new_wage_data)
plt.subplot(2,2,3)
plt.boxplot(wage_data,vert=False)
plt.subplot(2,2,4)
plt.boxplot(new_wage_data,vert=False)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.subplot(2,2,1).hist(wage_data)
plt.subplot(2,2,2).hist(new_wage_data)
plt.subplot(2,2,3).boxplot(wage_data,vert=False)
plt.subplot(2,2,4).boxplot(new_wage_data,vert=False)
plt.show()
```

**winsoriaztion**

- whatever the outliers less than lb fill with lb

- whatever the outliers greater than ub fill with ub

- in numpy we have method **clip**



```python
# step-1: get the wage data values
# step-2: calculate lb 
# step-3: calculate ub
# step-: chcek the parameters and apply in numpy np.clip()
# step-5: again implement above visulaization analysis 
90
wage_data=visa_df['prevailing_wage']
q1 = np.percentile(wage_data, 25)
q2 = np.percentile(wage_data, 50)
q3 = np.percentile(wage_data, 75)
iqr = q3 - q1
lb = q1 - 1.5 * iqr
ub = q3 + 1.5 * iqr
wage_data_clip=np.clip(wage_data,lb,ub)
```

```python
plt.figure(figsize=(10,6))
plt.subplot(2,2,1).hist(wage_data)
plt.subplot(2,2,2).hist(wage_data_clip)
plt.subplot(2,2,3).boxplot(wage_data,vert=False)
plt.subplot(2,2,4).boxplot(wage_data_clip,vert=False)
plt.show()
```
