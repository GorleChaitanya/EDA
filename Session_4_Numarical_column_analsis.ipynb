**Empirical Rule**

```python
#Step-1:
Mean=np.mean(pwage)
Mean
```

```python
Std=np.std(pwage)
Std
```

```python
lb=Mean-1*Std
lb
```

```python
ub=Mean+1*Std
ub
```

```python
con1=pwage>lb
con2=pwage<ub
con=con1&con2
len(visa_df[con])
```

```python
mean1=np.mean(pwage)
Std1=np.std(pwage)
lb1=mean1-2*Std1
ub1=mean1+2*Std1
con11=pwage>lb1
con22=pwage<ub1
conn=con11&con22
len(visa_df[conn])==24580*95/100
```
