**inplace**

- when we do the operation the result output is not overwriting

- means the original dataframe is same as it is.

- in order to do this we need use **inplace**

- by default inplace = False

- make **inplace=True** if you want overwrite the result

```python
df.drop('marks',axis=1,inplace=True)
```
